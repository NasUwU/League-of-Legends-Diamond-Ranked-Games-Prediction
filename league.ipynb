**Data Pre-processing and Cleaning**

```python
#displaying the column names along with the data types
data.info()
```

```python
data = data.drop("gameId", axis = 1)
```

```python
data.describe()
```

**Data Cleaning**

We can remove some columns that convey the same meaning.
1. Blue deaths = Red kills
2. Red deaths = Blue kills
3. Remove redGoldDiff as blueGoldDiff is still there
4. Remove redExperienceDiff as blueExperienceDiff is still there
5. Remove redFirstBlood because only one team can have the furst blood and for that blueFirstBlood is there

```python
to_drop = [
    "blueDeaths",
    "redDeaths",
    "redGoldDiff",
    "redExperienceDiff",
    "redFirstBlood"
]

data2 = data.drop(to_drop, axis = 1)
```

```python
#checking if the columns were removed
data2.head()
```

```python
#checking the number of columns after removing 
data2.shape
```

Missing Value Detection and Treatment

```python
#since this is Riot's official data, we can see here that there is no missing value
print(data2.isna().sum())
```

**Data Visualization**

```python
f, ax = plt.subplots(figsize=(50, 25))
corr = data2.corr("pearson")
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.dark_palette("#69d", reverse=True, as_cmap=True),
            square=True, ax=ax,annot=True)
```

```python
total=float(len(data))
plt.figure(figsize=(4, 3))
ax = sns.countplot(x="blueWins", data=data) 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2,height + 3,'{0:.0%}'.format(height/total),ha="center")
```

**Prediction**

Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X, y = data2.drop(["blueWins"], axis=1), data2["blueWins"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)
print(f"Training accuracy : {LRmodel.score(X_train, y_train)}")
print(f"Testing accuracy : {LRmodel.score(X_test, y_test)}")
```

```python
LRmodel.fit(X_train, y_train)
predicted= LRmodel.predict(X_test)
accuracy= LRmodel.score(X_test,y_test)
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Logistic Regression: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Logistic regression model: ', accuracy*100,'%')
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
GNBmodel = GaussianNB()
```

```python
GNBmodel.fit(X_train, y_train)
predicted= GNBmodel.predict(X_test)
accuracy= GNBmodel.score(X_test,y_test)
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Gaussian Naive Bayes Model: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Gaussian Naive Bayes Model: ', accuracy*100,'%')
```

XGBoost

```python
from xgboost import XGBClassifier
XGBmodel = XGBClassifier()
```

```python
XGBmodel.fit(X_train, y_train)
predicted = XGBmodel.predict(X_test)
accuracy = accuracy_score(y_test, predicted)
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Extreme Gradient Boosting Model: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Extreme Gradient Boosting Model: ', accuracy*100,'%')
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RFmodel = RandomForestClassifier()
```

```python
RFmodel.fit(X_train, y_train)
predicted = RFmodel.predict(X_test)
accuracy = accuracy_score(y_test, predicted)
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Random Forest Model: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Random Forest Model: ', accuracy*100,'%')
```
