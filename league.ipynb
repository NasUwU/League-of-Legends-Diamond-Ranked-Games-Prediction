**Data Pre-processing and Cleaning**

```python
#displaying the column names along with the data types
data.info()
```

```python
#dropping the column of gameId 
data = data.drop("gameId", axis = 1)
```

```python
data.describe()
```

**Data Cleaning**

We can remove some columns that convey the same meaning.
1. Blue deaths = Red kills
2. Red deaths = Blue kills
3. Remove redGoldDiff as blueGoldDiff is still there
4. Remove redExperienceDiff as blueExperienceDiff is still there
5. Remove redFirstBlood because only one team can have the first blood and for that blueFirstBlood is there

```python
to_drop = [
    "blueDeaths",
    "redDeaths",
    "redGoldDiff",
    "redExperienceDiff",
    "redFirstBlood"
]

data2 = data.drop(to_drop, axis = 1)
```

```python
#checking if the columns were removed
data2.head()
```

```python
#checking the number of columns after removing 
data2.shape
```

**Prediction**

Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X, y = data2.drop(["blueWins"], axis=1), data2["blueWins"]
```

70-30 split

```python
#we will be doing both the 70-30 and 80-20 split to observe if there is any difference in the accuracy of a model when the amount of data is increased
#here we are doing the 70-30 split at first
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)

predicted= LRmodel.predict(X_test)
accuracy= LRmodel.score(X_test,y_test)
accLR1 = accuracy
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Logistic Regression after doing the 70-30 split: ", accuracy) #previously used - metrics.accuracy_score(y_test, predicted) instead of accuracy
print('Accuracy score using the Logistic regression model after doing the 70-30 split: ', accuracy*100,'%')
```

80-20 split

```python
#now we will do the 80-20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=32)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)

predicted= LRmodel.predict(X_test)
accuracy= LRmodel.score(X_test,y_test)
accLR2 = accuracy
from sklearn import metrics
cm2 = metrics.confusion_matrix(y_test, predicted)
print(cm2)
```

```python
print("Accuracy of Logistic Regression after doing the 80-20 split: ",accuracy)
print('Accuracy score using the Logistic regression model after doing the 80-20 split: ', accuracy*100,'%')
```

```python
#printing classification report for Logistic Regression Model
from sklearn.metrics import classification_report

LRmodel = LogisticRegression()
LRmodel_pred = LRmodel.fit(X_train, y_train).predict(X_test)
target_names = ['Blue Loses', 'Blue Wins']

print(classification_report(y_test, LRmodel_pred, target_names=target_names))
```

```python
clf_report=classification_report(y_test, LRmodel_pred, output_dict=True)
plt.figure(figsize=(6, 5))
sns.heatmap(pd.DataFrame(clf_report).iloc[:-1, :].T, annot=True,  cmap="Blues")
```

```python
from sklearn.metrics import confusion_matrix
array = confusion_matrix(y_test, LRmodel_pred)
data2_cm2 = pd.DataFrame(array, index = ["Wins","Loses"],
                  columns = ["Wins","Loses"])
plt.figure(figsize = (5,4))
cm2_plot = sns.heatmap(data2_cm2, annot=True, cmap='Blues', fmt='g')
cm2_plot.set_xlabel('Predicted Labels')
cm2_plot.set_ylabel('True Labels')
```
