**Data Pre-processing and Cleaning**

```python
#displaying the column names along with the data types
data.info()
```

```python
data = data.drop("gameId", axis = 1)
```

```python
data.columns
```

```python
data.describe()
```

**Data Cleaning**

We can remove some columns that convey the same meaning.
1. Blue deaths = Red kills
2. Red deaths = Blue kills
3. Remove redGoldDiff as blueGoldDiff is still there
4. Remove redExperienceDiff as blueExperienceDiff is still there
5. Remove redFirstBlood because only one team can have the furst blood and for that blueFirstBlood is there

```python
to_drop = [
    "blueDeaths",
    "redDeaths",
    "redGoldDiff",
    "redExperienceDiff",
    "redFirstBlood"
]

data2 = data.drop(to_drop, axis = 1)
```

```python
#checking if the columns were removed
data2.head()
```

```python
#checking the number of columns after removing 
data2.shape
```

Missing Value Detection and Treatment

```python
#since this is Riot's official data, we can see here that there is no missing value
print(data2.isna().sum())
```

**Data Visualization** 

```python
data2.corr()["blueWins"].abs().sort_values(ascending=False)
```

```python
f, ax = plt.subplots(figsize=(50, 25))
corr = data2.corr("pearson")
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.dark_palette("#69d", reverse=True, as_cmap=True),
            square=True, ax=ax,annot=True)
```

```python
sns.set(rc={'figure.figsize':(15,15)})
corr = data2.corr()
sns.heatmap(corr[((corr >= 0.3) | (corr <= -0.3)) & (corr != 1)], annot=False, linewidths=.5, fmt= '.2f')
plt.title('Corelation Matrix');
```

```python
#here we are trying to see the distribution of data and we see how it is equally distributed and we have a an equal number of wins and losses in this dataset. A value of 1 means the blue team has won.
total=float(len(data))
plt.figure(figsize=(4, 3))
ax = sns.countplot(x="blueWins", data=data) 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2,height + 3,'{0:.0%}'.format(height/total),ha="center")
```

**Prediction**

Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X, y = data2.drop(["blueWins"], axis=1), data2["blueWins"]
```

```python
#we will be doing both the 70-30 and 80-20 split to observe if there is any difference in the accuracy of a model when the amount of data is increased
#here we are doing the 70-30 split at first
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)

predicted= LRmodel.predict(X_test)
accuracy= LRmodel.score(X_test,y_test)
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

```python
print("Accuracy of Logistic Regression: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Logistic regression model: ', accuracy*100,'%')
```

```python
#now we will do the 80-20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=32)
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)

predicted= LRmodel.predict(X_test)
accuracy= LRmodel.score(X_test,y_test)
from sklearn import metrics
cm2 = metrics.confusion_matrix(y_test, predicted)
print(cm2)
```

```python
print("Accuracy of Logistic Regression: ",metrics.accuracy_score(y_test, predicted))
print('Accuracy score using the Logistic regression model: ', accuracy*100,'%')
```

```python
from sklearn.metrics import classification_report

LRmodel = LogisticRegression()
LRmodel_pred = LRmodel.fit(X_train, y_train).predict(X_test)

print(classification_report(y_test, LRmodel_pred))
```

```python
from sklearn.metrics import classification_report

LRmodel = LogisticRegression()
LRmodel_pred = LRmodel.fit(X_train, y_train).predict(X_test)

target_names = ['Blue Loses', 'Blue Wins']
print(classification_report(y_test, LRmodel_pred, target_names=target_names))
```

```python
clf_report=classification_report(y_test, LRmodel_pred, output_dict=True)
plt.figure(figsize=(6, 5))
sns.heatmap(pd.DataFrame(clf_report).iloc[:-1, :].T, annot=True,  cmap="Blues")
```
